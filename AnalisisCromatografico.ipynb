Determinar a partir de la fraccion peso, la composicion molar y peso molecular de la muestra obtenida de un analisis cromatografico

```python
import  pandas as pd 
import numpy as np
```

### Composicion cromatografica

```python
DF = pd.read_csv("C:/Users/Luis Navarro/Desktop/ProgrammingStuff/PetroleumData/Ing.Yac.Gas/T06Datos.csv",encoding = 'unicode_escape')
DF.drop(DF.tail(2).index,inplace=True)
DF
```

```python
Tabla = {'CarbonFraction':['N2','CO2','C1','C2','C3','iC4','nC4','iC5','iC5','nC5','C6','C7','C8','C9','C10']}
Res = pd.DataFrame(Tabla,columns=['CarbonFraction'])
```

- Comprobacion de la fraccion total del peso 
$$ \sum{w_i}= 100\% $$
- Obtencion del producto de fraccion peso y el peso molecular
$$ w_i*{PM} $$

```python
print("sumatoria: ",sum(DF["Weight%"]))
DF["Wi/PM"] = DF["Weight%"]/DF["PM (g/mol)"]
DF
```

```python
N2_w = 1.6542
C02_w = 2.3040
C1_w = 60.5818
C2_w = 15.5326
C3_w = 12.3819
iC4_w = 2.0616
nC4_w = 3.2129
iC5_w = 0.0074
iC5 = 0.7677
nC5 = 0.6601
C6_w = DF.iloc[10:16,2].sum()
C7_w = DF.iloc[16:30,2].sum()
C8_w = DF.iloc[30:40,2].sum()
C9_w = DF.iloc[40:43,2].sum()
C10_w = DF.iloc[43,2]
```

- Obtencion de la sumatoria de fraccion sobre el peso
$$ \sum{\frac{w_i}{PM}} $$

```python
weight = [N2_w,C02_w,C1_w,C2_w,C3_w,iC4_w,nC4_w,iC5_w,iC5,nC5,C6_w,C7_w,C8_w,C9_w,C10_w ]
Res['Peso%'] = weight
sum_wp = sum(DF["Weight%"]/DF["PM (g/mol)"])
```

- Construccion de tabla de valores

```python
Res
```

- Determinacion de peso molecular $ M $ 
$$ M_{C_N} = \frac{w_{C_n}}{\sum{\frac{w_k}{M_k}}} $$

```python
N2_m = Res.iloc[0,1]/(N2_w/DF.iloc[0,3])
C02_m = Res.iloc[1,1]/(C02_w/DF.iloc[1,3])
C1_m = Res.iloc[2,1]/(C1_w/DF.iloc[2,3])
C2_m = Res.iloc[3,1]/(C2_w/DF.iloc[3,3])
C3_m = Res.iloc[4,1]/(C3_w/DF.iloc[4,3])
iC4_m = Res.iloc[5,1]/(iC4_w/DF.iloc[5,3])
nC4_m = Res.iloc[6,1]/(nC4_w/DF.iloc[6,3])
iC5_m = Res.iloc[7,1]/(iC5_w/DF.iloc[7,3])
C5_m = Res.iloc[8,1]/(iC5/DF.iloc[8,3])
nC5_m = Res.iloc[9,1]/(nC5/DF.iloc[9,3])
C6_m = Res.iloc[10,1]/DF.iloc[10:16,6].sum()
C7_m = Res.iloc[11,1]/DF.iloc[16:30,6].sum()
C8_m = Res.iloc[12,1]/DF.iloc[30:40,6].sum()
C9_m = Res.iloc[13,1]/DF.iloc[40:43,6].sum()
C10_m = Res.iloc[14,1]/DF.iloc[43,6]
```

```python
Res["M(g/mol)"]=[N2_m,C02_m,C1_m,C2_m,C3_m,iC4_m,nC4_m,iC5_m,C5_m,nC5_m,C6_m,C7_m,C8_m,C9_m,C10_m]
```

```python
Res
```

- Calculo de la fraccion molar de cada componente 
$$ Z_i = \frac{\frac{w_i}{PM_i}}{\sum{\frac{w_j}{PM_j}}} $$ 

```python
Zi = []
i = 0
for i in range(10):
    res=(DF.iloc[i,2]/DF.iloc[i,3])/sum_wp*100
    Zi.append(res)
    i = i+1
```

```python
Zi
```

```python
for i in range(5):
    res=(Res.iloc[i+10,1]/Res.iloc[i+10,2])/sum_wp*100
    Zi.append(res)
    i=i+1
```

```python
Zi
```

```python
Res["%Fraccion mol"] = Zi
```

```python
Res
```

- Comprobacion de fraccion molar
$$ \sum{Z_i} = 100\%$$
